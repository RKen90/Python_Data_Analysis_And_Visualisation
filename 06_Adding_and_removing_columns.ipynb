## Dropping columns

dataframe.drop()

```python
btc
```

```python
# Dropping the Symbol column
# Using axis = 1

btc.drop('Symbol', axis=1)
```

```python
# Using Axis = Columns

btc.drop('Symbol', axis = "columns")
```

## Dropping Rows

```python
countries.set_index('Country name', inplace=True)
```

```python
# dropping a row with this label
countries.drop(labels = 'Denmark', axis = 0)
```

```python
country = ['Finland', 'Iceland', 'Lesotho']
countries.drop(labels = country, axis = 0)
```

## Deleting Positionally

```python
# Deleting Positionally using a slice

countries.drop(countries.index[0:3])
```
