## Masking Updates with loc[ ] and Boolean Masks

```python
houses = pd.read_csv('CourseMaterials/data/kc_house_data.csv')
```

```python
houses
```

```python
too_many = houses['bedrooms'] >= 10
```

```python
too_many
```

```python
houses.loc[too_many, ['bedrooms', 'bathrooms']] =
```

```python
houses.loc[too_many, ['bedrooms']] = 'Too Many'
```

```python
houses[too_many]
```

```python
high_quality = houses['grade'] > 12
great_views = houses['view'] == 4
```

```python
houses[great_views & high_quality]
```

```python
houses['luxury level'] = 'normal'
```

```python
houses
```

```python
houses.loc[great_views & high_quality, ['luxury_level']] = 'ultra'
```

```python
houses
```
