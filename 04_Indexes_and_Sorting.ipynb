## Sorting by multiple columns

## loc

- Helps us access data by row

```python
happy
```

```python
happy.loc[['Brazil', 'Egypt', 'United Kingdom']]
```

```python
happy.sort_index().loc["Denmark":"France"]
```

## iloc

- Purely integer-location based indexing for selection by position

```python
happy.iloc[[20]]
```

```python
happy.iloc[[10, 20, 30]]
```

```python
happy.iloc[50:60]
```

```python
houses.sort_index(ascending=False)
```
