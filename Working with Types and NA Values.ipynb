## Converting objects to numeric data
- using astype() method

```python
titanic['age'].value_counts()
```

```python
titanic['age'].replace(['?'], [None], inplace = True)
```

```python
titanic['age'].astype('float')
```

```python
titanic['age_float'] = titanic['age'].astype('float')
```

```python
titanic
```

```python
titanic['age'].astype('float').mean()
```

```python
titanic['age_float'].mean()
```

```python
titanic.info()
```

## Category Type

- Represents text data where there is a finite number of possibilities e.g. M or F
- Reduces the overall size of the dataset

```python
titanic['sex'].astype("category")
```

```python
titanic['sex'] = titanic['sex'].astype("category")
```

```python
titanic.info()
```

```python
titanic['embarked'] = titanic['embarked'].astype('category')
```

```python
titanic.info()
```

 ## Casting with .to_numeric( )
