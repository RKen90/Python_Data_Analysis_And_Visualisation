## Combining Conditions using AND (&)

```python
titanic
```

```python
women = titanic['sex'] == 'female'
died = titanic['survived'] == 0
titanic[women & died]
```

```python
titanic[(titanic['sex'] == 'female') & (titanic['survived'] == 0)]
```

```python
houses[(houses['waterfront'] == 1) & (houses["price"] < 500000)]
```

```python
houses[(houses['view'] == 4) & (houses['grade'] >= 10)]
```

## Combining conditions using OR(|)

True | False

```python
#homes built after 2010 or homes renovated after 2010

houses
```

```python
built = houses['yr_built'] >= 2014
renovated = houses['yr_renovated'] >= 2014

houses[built | renovated]
```

```python
netflix
```

```python
netflix[(netflix['director'] == 'David Fincher') | (netflix['director'] == 'Martin Scorsese')]
```

```python
fincher = netflix['director'] == 'David Fincher'
scorsese = netflix['director'] == 'Martin Scorsese'
released = netflix['release_year'] > 2015
netflix[(fincher | scorsese) & (released)]
```
