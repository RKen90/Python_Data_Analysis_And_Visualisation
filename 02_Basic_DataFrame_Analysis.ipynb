# Count

- The `count` function in pandas is used to **calculate the number of non-NA/null entries across each column or row of a DataFrame or Series**. 
- This function helps identify how many valid entries are present, which is essential for data analysis to ensure the integrity and completeness of the data set.

# Mean, Median and Mode


## Mean
- The Average 
    - The sum of all the numbers divided by the amount of numbers
    
    
## Median
- The middle number
    - The numbers are organised, the number in the middle is the Median
    
    
## Mode
- The most occuring number

```python
houses.price
```

# Describe with Numeric Values

## Generates Descriptive Statistics

- The `describe` function in pandas **provides a summary of statistics pertaining to DataFrame or Series columns**. - This includes central tendency, dispersion, and shape of the dataset’s distribution, excluding NaN values. 
- Typically, it returns information such as mean, median, mode, minimum, maximum, standard deviation, and count for numeric data.
- For object data (e.g., strings or timestamps), it provides count, unique, top, and freq (the most common value's frequency). This function is invaluable for quickly understanding and summarizing the data's distribution and central characteristics.

```python
titanic.describe()
```

```python
titanic.survived.describe()
```

```python
houses.describe()
```

```python
houses.price.describe()
```
